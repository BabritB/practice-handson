```

import pandas as pd
from googlesearch import search 
def search_google(query):
    parent_=[]  
    for j in search(query, tld="co.in", num=10, stop=50, pause=2):
        child_=[]
        link_=j
        site_name=link_.split("/")[2]
        ss=link_.split("/")
        child_.append(site_name)
        child_.append(link_)
        if '.' in ss[len(ss)-2]:child_.append('')
        else:child_.append(ss[len(ss)-2])
        parent_.append(child_) 
    return pd.DataFrame(parent_,columns=['Site Name','URL Link','Name'])
    #return parent_
```

```
search_google('python')
```
